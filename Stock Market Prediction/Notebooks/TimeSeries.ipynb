```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import investpy
```

```python
# list of all the stocks present in india
stock_list = investpy.stocks.get_stocks_list("india")
print("Num Stocks: ", len(stock_list))
```

```python
# Get Stocks
stock_list = investpy.stocks.get_stocks("india")
stock_list.head()
```

```python
# Stocks Overview
stocks_overview = investpy.stocks.get_stocks_overview("india")
```

```python
# Search for stock
search_result = investpy.search_quotes(text="RELIANCE", products =["stocks"], countries=["india"], n_results = 1)
print("Name: ", search_result.name)
print("Symbol: ", search_result.symbol)
```

```python
# Get Stock information 
investpy.stocks.get_stock_information("RELI", "india")
```

```python
# Get Stock Historic Data
df_hist = investpy.stocks.get_stock_historical_data(stock = "RELI", country = "india", 
                                          from_date = "1/1/2020", 
                                          to_date = "1/1/2021", 
                                          order = 'ascending', 
                                          interval = 'Daily' # Weekly or Monthly
                                         )
df_hist.head(2)
```

```python
# Get stock summary financial (quartarly and annually)
stock_summary = investpy.stocks.get_stock_financial_summary(stock = "RELI",
                                                            country = "india",
                                                            summary_type = "income_statement", # cash_flow_statement or balance_sheet
                                                            period = "annual" # quarterly
                                                            )
stock_summary.head()
```

```python
# Get stock dividends
investpy.stocks.get_stock_dividends(stock = "RELI", 
                                    country = 'india')
```

```python
# Get company Profile
investpy.stocks.get_stock_company_profile("RELI", "india")
```
